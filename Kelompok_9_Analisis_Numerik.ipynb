Runge-Kutta (RK4) is a simple method for numerical analysis like Euler which requires an initial value. Runge-Kutta has a relatively lower error than Euler because it uses the average slope, while Euler's accuracy depends on how big h (step size) is.

```python
import matplotlib.pyplot as plt
import numpy as np
from scipy.integrate import solve_ivp
```

```python
def runge(x, y):
    return x**2 - y

# Step size iterasi
h = 0.01
# Initial condition
x = 0
y = 0

Xres = [x]
Yres = [y]

while x <= 1:
    k1 = h * runge(x, y)
    k2 = h * runge(x + h/2, y + k1/2)
    k3 = h * runge(x + h/2, y + k2/2)
    k4 = h * runge(x + h, y + k3)

    y = y + (1/6)*(k1 + 2*k2 + 2*k3 + k4)
    x = x + h

    Xres.append(x)
    Yres.append(y)

plt.plot(Xres, Yres)
plt.title("Runge Kutta Method")
plt.xlabel("x")
plt.ylabel("y")
plt.show()
```

```python
import matplotlib.pyplot as plt

def f(x, y):
    return x**2 - y

h = 0.01
x = 0
y = 0

X = [x]
Y = [y]

# List untuk menyimpan titik k1–k4
kx = {"k1": [], "k2": [], "k3": [], "k4": []}
ky = {"k1": [], "k2": [], "k3": [], "k4": []}

step_show = 5   # tampilkan titik hanya untuk 5 langkah pertama
step = 0

while x <= 1:
    k1 = h * f(x, y)
    k2 = h * f(x + h/2, y + k1/2)
    k3 = h * f(x + h/2, y + k2/2)
    k4 = h * f(x + h, y + k3)

    # Simpan titik k1–k4 untuk visualisasi
    if step < step_show:
        kx["k1"].append(x)
        ky["k1"].append(y)

        kx["k2"].append(x + h/2)
        ky["k2"].append(y + k1/2)

        kx["k3"].append(x + h/2)
        ky["k3"].append(y + k2/2)

        kx["k4"].append(x + h)
        ky["k4"].append(y + k3)

    y = y + (1/6)*(k1 + 2*k2 + 2*k3 + k4)
    x = x + h

    X.append(x)
    Y.append(y)

    step += 1


# =========================
#  PLOTTING
# =========================
plt.figure(figsize=(8,5))
plt.plot(X, Y, label="Solusi RK4", linewidth=2)

# Plot titik k1-k4
plt.scatter(kx["k1"], ky["k1"], color='red', label='k1', s=25)
plt.scatter(kx["k2"], ky["k2"], color='blue', label='k2', s=25)
plt.scatter(kx["k3"], ky["k3"], color='green', label='k3', s=25)
plt.scatter(kx["k4"], ky["k4"], color='purple', label='k4', s=25)

plt.title("Visualisasi Titik k1, k2, k3, k4 pada Runge–Kutta 4")
plt.xlabel("x")
plt.ylabel("y")
plt.legend()
plt.grid(True)
plt.show()
```

```python
F = lambda t, s: np.dot(np.array([[0, t**2], [-t, 0]]), s)

t_eval = np.arange(0, 10.01, 0.01)
sol = solve_ivp(F, [0, 10], [1, 1], t_eval=t_eval)

plt.figure(figsize = (12, 8))
plt.plot(sol.y.T[:, 0], sol.y.T[:, 1])
plt.xlabel('x')
plt.ylabel('y')
plt.show()
```
